```python
import pandas as pd
import numpy as np
import ipdb
```

```python
dflist = [pd.read_excel('sim_matrix_cos.xlsx',sheet_name=f'Q{i}') for i in range(1,14)]
```

```python
df_rank_list = []
for iSheet,df in enumerate(dflist,start=1):
    words = df.iloc[:,0]
    coldict = {}
    for icol in [1,2,3,4,5]:
        col = df.loc[:,df.columns==icol].values
        coldict[icol]=col
    df_rank = pd.DataFrame({key:item.ravel() for key,item in coldict.items()},index=words)
    if iSheet==1:
        df_rank_list=df_rank
        #ipdb.set_trace()
    else:
        #display(df_rank)
        df_rank_list = df_rank_list.append(df_rank)
        #print(iSheet,df.shape,df_rank_list.shape)
    print(df_rank_list.shape)
    display(df_rank_list.tail(3))

```
